```python
# 上述两个数据集，我已经下载好解压放入同级目录下，方便访问

# print('上述两个数据集，我已经下载好解压放入同级目录下，方便访问'.encode('utf-8').decode('gb18030'))
print(thu1.cut("上述两个数据集，我已经下载好解压放入同级目录下，方便访问", text=True))

def getCsvArticles(filename, contentColName, encoding='gb18030', toEncoding = 'utf-8'):
    content = pd.read_csv(filename, encoding=encoding)
    articles = content[contentColName].tolist()
    return [article for i, article in enumerate(articles)  if isinstance(article, str) ]

def save(wordList, fileName, endSplit = "\n"):
    with open(fileName, "w", encoding='utf-8') as f:
        for a in wordList:
            f.write(a+endSplit)
```

```python
endSplit = '\n\n\n\n\n\n\n\n\n\n\n\n'

articles1 = getCsvArticles('OriginalData/sqlResult_1558435.csv', 'content')
print(articles1[0])
news_articles_data_path = 'NewsArticlesData/'

thu1 = thulac.thulac(user_dict=None, model_path=None, T2S=True, seg_only=True) 

articles2 = []
for i, article in enumerate(articles1):
    articles2.append(thu1.cut(article, text = True))
    if i% 10 == 9:
        print('已完成：', i + 1, '/', len(articles1))
print(articles2)
print("全部完成！")

```

```python
print(articles2[0])
save(articles2, news_articles_data_path + '/sqlResult_News_cut.txt', endSplit)
```
