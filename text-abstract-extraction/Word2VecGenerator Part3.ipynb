## Word2Vect 模型训练

```python
# https://blog.csdn.net/qq_27586341/article/details/90025288
from gensim.models import word2vec, Word2Vec 
import os
```

```python
news_articles_data = 'NewsArticlesData/sqlResult_News_cut.txt'
# news_articles_data = 'WikiArticlesData/wiki_article_0.txt'
```

```python
def analogy(model, x1, x2, y1):
    try:
        result = model.wv.most_similar(positive = [y1,x2],negative = [x1])
        return '%s,%s  %s->%s'%(x1, x2, y1, result[0][0])
    except KeyError as keyError:
        print(keyError)
        return None
    
def most_similar(model, x1):
    try:
        result = model.wv.most_similar(x1)
        return '%s -> %s'%(x1, result)
    except KeyError as keyError:
        print(keyError)
        return None
    
def printTest(model):
    print(model)
    print(analogy(model, "中国","汉语","美国"))
    print(analogy(model, "美国", "奥巴马", "日本"))
    print(most_similar(model, '勇敢'))
    print(most_similar(model, '美女'))
```

```python

def word2vec_train(articles_data_path:str, model:Word2Vec=None,size=50, window=5):
    sentences = word2vec.LineSentence(articles_data_path)

    input_sentences = []
    # 训练一个批次的指针
    count = 0
    # 每次训练的训练量
    batch_size = 5000
    for i, sentence in enumerate(sentences):
        input_sentences.append(sentence)
        # 每20万行训练一次，之后将input_sentences清空
        if (i + 1) % batch_size == 0:
            if model is None:
                model = Word2Vec(input_sentences, size=size,window=window)
                input_sentences = []
                count += 1
                print("成功训练{}次模型".format(count))
            else:
                #增量训练
                model.build_vocab(input_sentences, update=True) #更新词汇表
                # epoch=iter语料库的迭代次数；（默认为5）  total_examples:句子数。
                model.train(input_sentences, total_examples=batch_size, epochs=model.iter)
                input_sentences = []
                count += 1
                print("成功训练{}次模型".format(count))
    if len(input_sentences) > 0:
        print(len(input_sentences))
        if model is None:
            model = Word2Vec(input_sentences, size=size,window=window)
        else:
            model.build_vocab(input_sentences, update=True) #更新词汇表
            model.train(input_sentences, total_examples=batch_size, epochs=model.iter)
        print("成功训练{}次模型".format(count))
        
    return model

```

```python
model = word2vec_train(news_articles_data)
```

```python
model.save('Data/word2vect_50_w5.model')
print('模型保存完毕！')
```

```python
model = Word2Vec.load('Data/word2vect_50_w5.model')
printTest(model)
```

```python
# wiki_articles_data = 'WikiArticlesData/wiki_article_%d.txt'
wiki_articles_data = 'WikiArticlesData/'

allFiles = os.listdir(wiki_articles_data)

allFileCount = len(allFiles)

for i, filePath in enumerate(allFiles):
    model = word2vec_train(wiki_articles_data + filePath, model, size=50, window=5)
    if i% 10 == 9:
        print('已完成：', i + 1, '/', allFileCount, model)
print("全部完成！")
```
